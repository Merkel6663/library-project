```python
def trap(f, a, b, tol):
    n = 1
    I1 = 0
    while True:
        x = np.linspace(a, b, n+1)
        y = f(x)
        h = (b-a)/n
        I2 = h*((y[0]+y[-1])/2 + np.sum(y[1:-1]))
        if np.abs(I2-I1) < tol:
            break
        n *= 2
        I1 = I2
    return [I2, n]


def rect(f, a, b, tol):
    n = 2
    I1 = 0
    while True:
        h = (b-a)/n
        x_mid = np.linspace(a+h/2, b-h/2, n)
        I2 = h * np.sum(f(x_mid))
        if np.abs(I2-I1) < tol:
            break
        n *= 2
        I1 = I2
    return [I2, n]

def simpson(f, a, b, tol):
    n = 2
    I1 = 0
    while True:
        x = np.linspace(a, b, n+1)
        y = f(x)
        h = (b-a)/n
        I2 = h/3*(y[0]+y[-1]+4*np.sum(y[1:-1:2])+2*np.sum(y[2:-2:2]))
        if np.abs(I2-I1) < tol:
            break
        n *= 2
        I1 = I2
    return [I2, n]



def f(x):
    y = (1 + x**2 + x**4)**0.5
    return y

import numpy as np
import matplotlib.pyplot as plt

a = -2
b = 3
tol = 10**(-8)
result1 = rect(f, a, b, tol)
print("\nРезультат метода прямоугольников:", result1[0], "\n Количсетво итераций:", result1[1])
result2 = trap(f, a, b, tol)
print("\nРезультат метода трапеций:", result2[0], "\n Количсетво итераций:", result2[1])
result3 = simpson(f, a, b, tol)
print("\nРезультат метода Симпсона:", result3[0], "\n Количсетво итераций:", result3[1])

x = np.linspace(a, b, max(result1[1], result2[1], result3[1])+1)
plt.plot(x, f(x),x,f(x)*0)
plt.plot([a,a],[f(a)*0,f(a)])
plt.plot([b,b],[f(b)*0,f(b)])
plt.grid(True)
```

```python
def trap(f, a, b, tol):
    n = 2
    I1 = 0
    while True:
        x = np.linspace(a, b, n+1)
        y = f(x)
        h = (b-a)/n
        I2 = h*((y[0]+y[-1])/2 + np.sum(y[1:-1]))
        if np.abs(I2-I1) < tol:
            break
        n *= 2
        I1 = I2
    return I2

def rect(f, a, b, tol):
    n = 2
    I1 = 0
    while True:
        h = (b-a)/n
        x_mid = np.linspace(a+h/2, b-h/2, n)
        I2 = h * np.sum(f(x_mid))
        if np.abs(I2-I1) < tol:
            break
        n *= 2
        I1 = I2
    return I2

def simpson(f, a, b, tol):
    n = 2
    I1 = 0
    while True:
        x = np.linspace(a, b, n+1)
        y = f(x)
        h = (b-a)/n
        I2 = h/3*(y[0]+y[-1]+4*np.sum(y[1:-1:2])+2*np.sum(y[2:-2:2]))
        if np.abs(I2-I1) < tol:
            break
        n *= 2
        I1 = I2
    return I2

import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return np.sqrt(x) * np.sin(x)

def f2(x):
    return np.log(5 + 4*np.cos(x))

def f3(x):
    sin_x = np.sin(x)
    sqrt_sin = np.sqrt(np.abs(sin_x)) # модуль для того чтобы избежать отрицательного значения под корнем на границе
    return sqrt_sin * np.cos(np.sqrt(x))

a1, b1 = 0, np.pi
a2, b2 = np.pi, 2*np.pi
a3, b3 = 2*np.pi, 2.8*np.pi
tol = 10**(-6)

rect_sum = rect(f1, a1, b1, tol) + rect(f2, a2, b2, tol) + rect(f3, a3, b3, tol)
trap_sum = trap(f1, a1, b1, tol) + trap(f2, a2, b2, tol) + trap(f3, a3, b3, 10**(-8))
simpson_sum = simpson(f1, a1, b1, tol) + simpson(f2, a2, b2, tol) + simpson(f3, a3, b3, tol)

print("\nРезультат метода прямоугольников:", rect_sum)
print("Результат метода трапеций:", trap_sum)
print("Результат метода Симпсона:", simpson_sum)

x1 = np.linspace(a1, b1, 100)
x2 = np.linspace(a2, b2, 100)
x3 = np.linspace(a3, b3, 100)

plt.plot(x1, f1(x1), 'b-')
plt.plot(x2, f2(x2), 'g-')
plt.plot(x3, f3(x3), 'r-')
plt.plot(x1, x1*0, 'k-')

plt.plot([a1, a1], [0, f1(a1)], 'r--')
plt.plot([b1, b1], [0, f1(b1)], 'r--')
plt.plot([b2, b2], [0, f2(b2)], 'r--')
plt.plot([b3, b3], [0, f3(b3)], 'r--')

plt.grid(True)
plt.show()
```
