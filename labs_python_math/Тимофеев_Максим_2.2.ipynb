```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
y1 = x * sin(x)
y2 = x**2 - 1



x = np.linspace(-10, 10, 400)
y1_vals = x * np.sin(x)
y2_vals = x**2 - 1


plt.figure(figsize=(10, 6))
plt.plot(x, y1_vals, label='y = x·sin(x)', linewidth=2)
plt.plot(x, y2_vals, label='y = x² - 1', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.title('Графики функций')
plt.show()
```
